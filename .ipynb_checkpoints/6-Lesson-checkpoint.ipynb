```python
# Fill with mean
df['Experience_Years'].fillna(df['Experience_Years'].mean(), inplace=True)

# Fill with median
df['Monthly_Salary_USD'].fillna(df['Monthly_Salary_USD'].median(), inplace=True)

# Fill with 0 if missing means “none” or “no experience”
df['Training_Hours'].fillna(0, inplace=True)

# Fill with the most repeated ones
df['Department'].fillna(df['Department'].mode()[0], inplace=True)

df['Remote_Work'].fillna('Unknown', inplace=True)
```

```python
df['Work_Hours_per_Week'].fillna(df['Work_Hours_per_Week'].median(), inplace=True)
df['Performance_Score'].fillna(df['Performance_Score'].mode()[0], inplace=True)
df.isnull().sum()
```
